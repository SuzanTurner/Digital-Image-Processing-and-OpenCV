```python
import cv2
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
image = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\SuzanTurner\19maib.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
mask = np.zeros(image.shape[:2], np.uint8)

bg_model = np.zeros((1, 65), np.float64)
fg_model = np.zeros((1, 65), np.float64)

height, width = image.shape[:2]
rect = (10, 10, width - 25, height - 25)  

cv2.grabCut(image, mask, rect, bg_model, fg_model, 5, cv2.GC_INIT_WITH_RECT)

mask_bin = np.where((mask == 2) | (mask == 0), 0, 1).astype("uint8")

foreground = image * mask_bin[:, :, np.newaxis]

# Display results
plt.subplot(1, 2,1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")

plt.subplot(1, 2,2)
plt.imshow(foreground)
plt.title("Background Removed")
plt.axis("off")
```
