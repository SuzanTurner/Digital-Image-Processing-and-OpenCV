```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image = cv2.imread(r"C:\Users\hp\OneDrive\Pictures\Screenshots\chess.png")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Harris Corner Detection
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

# Dilate corner points for better visualization
dst = cv2.dilate(dst, None)

# Threshold to mark the corners
image[dst > 0.01 * dst.max()] = [170, 255, 0]  #Marks corners in green

# Convert image to RGB for displaying in Matplotlib
image_copy = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
plt.axis("off")  
plt.title("Harris Corner Detection")
plt.show()
```
