### Transelation

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def translate_image(input_image_path, tx, ty):
    img = Image.open(input_image_path)
    img_array = np.array(img)
    rows, cols, _ = img_array.shape

    translated_img = np.zeros_like(img_array)

    for i in range(rows):
        for j in range(cols):
            new_i, new_j = i + ty, j + tx
            if 0 <= new_i < rows and 0 <= new_j < cols:
                translated_img[new_i, new_j] = img_array[i, j]

    translated_img = Image.fromarray(translated_img)
    plt.imshow(translated_img)
    plt.axis('off')
    plt.title('Translated Image')
    plt.show()
translate_image(r'D:\Yadh Documents\College Stuff\Pictures\Silver Void.jpg', tx=50, ty=50)
```

### Rotation

```python
from PIL import Image
import matplotlib.pyplot as plt

def rotate_image(input_image_path, angle):
    img = Image.open(input_image_path)
    
    rotated_img = img.rotate(angle, resample=Image.BICUBIC, expand=True)
    
    plt.imshow(rotated_img)
    plt.axis('off')
    plt.title(f'Rotated Image by {angle}°')
    plt.show()

rotate_image(r'D:\Yadh Documents\College Stuff\Pictures\Silver Void.jpg', angle=69)
```

### Scaling

```python
from PIL import Image
import matplotlib.pyplot as plt

def scale_image(input_image_path, scale_x, scale_y):
    img = Image.open(input_image_path)
    width, height = img.size
    new_width = int(width * scale_x)
    new_height = int(height * scale_y)

    scaled_img = img.resize((new_width, new_height), Image.BICUBIC)

    plt.imshow(scaled_img)
    plt.axis('off')
    plt.title(f'Scaled Image ({scale_x}x, {scale_y}x)')
    plt.show()

scale_image(r'D:\Yadh Documents\College Stuff\Pictures\Silver Void.jpg', scale_x=1.5, scale_y=0.5)
```

### Shearing

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def shear_image(input_image_path, shear_x, shear_y):
    img = Image.open(input_image_path)
    img_array = np.array(img)
    rows, cols, channels = img_array.shape

    # Shear transformation matrix
    shear_matrix = np.array([[1, shear_x, 0],
                             [shear_y, 1, 0],
                             [0, 0, 1]])

    # Create an empty image
    sheared_img = np.zeros_like(img_array)

    # Apply the shearing
    for i in range(rows):
        for j in range(cols):
            new_coords = np.dot(shear_matrix, np.array([j, i, 1]))
            new_j, new_i = int(new_coords[0]), int(new_coords[1])
            if 0 <= new_i < rows and 0 <= new_j < cols:
                sheared_img[new_i, new_j] = img_array[i, j]

    # Convert back to an image and display
    sheared_img = Image.fromarray(sheared_img)
    plt.imshow(sheared_img)
    plt.axis('off')
    plt.title('Sheared Image')
    plt.show()

# Replace with your image path and shearing values
shear_image(r'D:\Yadh Documents\College Stuff\Pictures\Silver Void.jpg', shear_x=0.2, shear_y=0.1)
```
