```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt

def load_image(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError("Image not found or invalid format.")
    return img

def display_image(title, img):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

image_path = r'D:\Yadh Documents\College Stuff\Pictures\Silver Void.jpg'
img = load_image(image_path)
```


### Thresholding

```python

def apply_threshold(img, threshold_value=128):
    _, thresh_img = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)
    display_image("Thresholded Image", thresh_img)
    return thresh_img

thresh_img = apply_threshold(img, threshold_value=128)
```

### Dilation

```python
def apply_dilation(img, kernel_size=(5, 5), iterations=1):
    kernel = np.ones(kernel_size, np.uint8)
    dilated_img = cv2.dilate(img, kernel, iterations=iterations)
    display_image("Dilated Image", dilated_img)
    return dilated_img

dilated_img = apply_dilation(thresh_img, kernel_size=(5, 5), iterations=2)
```

### Erosion

```python

def apply_erosion(img, kernel_size=(5, 5), iterations=1):
    kernel = np.ones(kernel_size, np.uint8)
    eroded_img = cv2.erode(img, kernel, iterations=iterations)
    display_image("Eroded Image", eroded_img)
    return eroded_img

eroded_img = apply_erosion(thresh_img, kernel_size=(5, 5), iterations=2)
```
